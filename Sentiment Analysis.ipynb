```python
def count_syllables(word):
    word = word.lower()
    if word.endswith(('es', 'ed')):
        return max(1, sum(1 for char in word if char in 'aeiou') - 1)#Subtracting 1 is a rule for words 
    #ending with 'es' or 'ed' because, in some cases, the 'e' is silent or not counted as a separate
    #syllable.
    else:
        return max(1, sum(1 for char in word if char in 'aeiou'))
```

```python
for index, row in df.iterrows():
    link = row['URL']

    try:
        # Fetch web content with headers
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
        response = requests.get(link, headers=headers)
        response.raise_for_status()  # Raise an exception for bad responses
    except requests.exceptions.RequestException as e:
        print(f"Error fetching content from {link}: {e}")
        continue  # Skip to the next row in case of an error

    html_content = response.text
    # Extract text from HTML
    soup = BeautifulSoup(html_content, 'html.parser')
    text = soup.get_text()

    # Tokenize and remove stop words
    #isalpha - is alphabetical or not
    words = word_tokenize(text)
    
    # Tokenize sentences
    sentences = sent_tokenize(text)
    

    #cleaning using stop words given
    filtered_words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]

    # Find positive score
    positive_score = sum(1 for word in filtered_words if word.lower() in positive_words)
    
    #Find Negative Score and multiplying it with -1
    negative_score = -1*sum(-1 for word in filtered_words if word.lower() in negative_words)
    
    #Calculating Polarity score using given formula
    #Range = -1 to +1
    polarity_score = (positive_score - negative_score)/((positive_score + negative_score + 0.000001))
    
    #Calculating Subjectivity_Score using given formula
    #Range = 0 to +1
    subjectivity_score = (positive_score + negative_score)/ (len(filtered_words)+ 0.000001)
    
    #Analysis of Readability
    #Finding Average Sentence Length 
    # Calculate average sentence length
    total_sentences = len(sentences)
    average_sentence_length = len(words) / total_sentences
    
    #Complex word count - words which contains more than 2 syllables

    complex_words = [word for word in words if count_syllables(word) > 2 and word.lower() not in stop_words]

    # Calculate the percentage of complex words
    percentage_complex_words = len(complex_words) / len(words) * 100
     
    #Gunning Fox index formula 
    fog_index = 0.4 * (average_sentence_length + percentage_complex_words )
    
    #Average Number of Words Per Sentence = the total number of words / the total number of sentences

    avg_no_of_words_per_sent = len(words) / total_sentences
    
    # Calculate the syllable count per word
    syllable_count_per_word = sum(count_syllables(word) for word in words) / len(words)
    
    # Calculate the Average Word Length
    total_characters_in_words = sum(len(word) for word in words)
    total_words = len(words)

    average_word_length = total_characters_in_words / total_words 

    # Update DataFrame with positive word count
    df.at[index, 'POSITIVE SCORE'] = positive_score
    df.at[index, 'NEGATIVE SCORE'] = negative_score
    df.at[index, 'POLARITY SCORE'] = polarity_score
    df.at[index, 'SUBJECTIVITY SCORE'] = subjectivity_score
    df.at[index, 'AVG SENTENCE LENGTH'] = average_sentence_length
    df.at[index, 'PERCENTAGE OF COMPLEX WORDS'] = percentage_complex_words
    df.at[index, 'FOG INDEX'] = fog_index
    df.at[index, 'AVG NUMBER OF WORDS PER SENTENCE'] = avg_no_of_words_per_sent
    df.at[index, 'COMPLEX WORD COUNT'] = len(complex_words)
    df.at[index, 'WORD COUNT'] = len(filtered_words)
    df.at[index, 'SYLLABLE PER WORD'] = syllable_count_per_word
    df.at[index, 'AVG WORD LENGTH'] = average_word_length

# Save the updated DataFrame to the same Excel file
df.to_csv(r"C:\Users\PRIYANKA\OneDrive\Desktop\File\Output Data Structure.xlsx", index=False)
```
